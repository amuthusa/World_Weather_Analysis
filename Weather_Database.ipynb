```python
#import dependency
import pandas as pd
import numpy as np
#import citipy module which would help to find city based on lattitude and longitude
from citipy import citipy
#import the open weather api key from config python file
from config import open_weather_api_key;
#import request to make http calls
import requests
#import datetime for datime conversion
from datetime import datetime
```

```python
#generate random latitude and longitude
latitudes = np.random.uniform(-90.000, 90.000, 1500)
longitudes = np.random.uniform(-180.000, 180.000, 1500)

#create tuples of latitude and longitude
lats_lngs = zip(latitudes, longitudes)
lats_lngs
```

```python
#create coordinate using the latitude longitude tuple
coordinates = list(lats_lngs)
coordinates
```

```python
#using the coordinate find the nearest citi using citipy
cities = []

for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    
    if (city not in cities):
        cities.append(city)

len(cities)
```

```python
#base url for openweather api to get weather data
base_url= "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + open_weather_api_key
```

```python
#list of city data
city_data = []

#start of data loading
print("----------------------------------")
print("Data Retrieval is started        ")
print("----------------------------------")


#enumerate over the cities list to get the index and city name 
for index, city in enumerate(cities):
    city_url = base_url + "&q="+city
    try:
        #make the get weather for the city as a json
        city_weather_json = requests.get(city_url).json()
         #Parse output and get the data needed
        city_lat = city_weather_json["coord"]["lat"]
        city_lng = city_weather_json["coord"]["lon"]
        city_max_temp = city_weather_json["main"]["temp_max"]
        city_humidity = city_weather_json["main"]["humidity"]
        city_clouds = city_weather_json["clouds"]["all"]
        city_winds = city_weather_json["wind"]["speed"]
        city_country = city_weather_json["sys"]["country"]
        city_weather_description = city_weather_json["weather"][0]["description"]
        #if its raining get the amount of rainfall in last 3h in inches else set it 0
        try:
            city_rainfall_inches = city_weather_json["rain"]["3h"]
        except KeyError as e:
            city_rainfall_inches = 0
        
        #same with snowfall get the amount of snowfall in last 3h in inches else set it 0
        try:
            city_snowfall_inches = city_weather_json["snow"]["3h"]
        except KeyError as e:
            city_snowfall_inches = 0;
        
        #convert date to iso standard
        city_date = datetime.utcfromtimestamp(city_weather_json["dt"]).strftime('%Y-%m-%d %H:%M:%S')
        
        #add all the data to city list
        city_data.append({"City": city.title(),
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_winds,
                          "Weather Description": city_weather_description,
                          "Country": city_country,
                          "Rainfall": city_rainfall_inches,
                          "Snowfall": city_snowfall_inches,
                          "Date": city_date})
    except KeyError as e:
        #check if the json response has message if so print
        if "message" in city_weather_json:
            print (city_weather_json["message"])
        else:
            print ("City not found")

#end of data loading
print("----------------------------------")
print("Data Retrieval is complete        ")
print("----------------------------------")

city_data
```

```python
#create a new dataframe using the city data
city_data_df = pd.DataFrame(city_data)
city_data_df.head(10)
```

```python
city_data_df = city_data_df.rename(columns={'Weather Description': 'Current Description', 'Rainfall': 'Rain inches (last 3 hrs)', 
                             'Snowfall': 'Snow inches (last 3 hrs)'})
```

```python
#create new Column Order
new_column_order = ["City", "Country", "Date", "Lat", "Lng", "Max Temp", "Humidity", "Cloudiness", 
                    "Wind Speed", "Current Description", "Rain inches (last 3 hrs)", "Snow inches (last 3 hrs)"]
city_data_df = city_data_df[new_column_order]

city_data_df
```

```python
#save the data as a csv file
output_file = "weather_data/WeatherPy_challenge.csv"
city_data_df.to_csv(output_file)
```
