```python
#import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#create a set of lattitude and longitude combinations
lats = np.random.uniform(-90.000, 90.000, 1500)
lngs = np.random.uniform(-180.000, 180.000, 1500)
lats_lngs = zip(lats, lngs)
lats_lngs
```

```python
#Add the latitudes and longitudes to a list
coordinates = list(lats_lngs)
coordinates
```

```python
#import citipy module which would help to find city based on lattitude and longitude
from citipy import citipy
```

```python
#iterate over the coordinates and find the nearest city using citipy
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1])
    print(city.city_name,",",city.country_code)
```
