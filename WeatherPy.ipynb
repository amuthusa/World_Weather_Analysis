```python
#import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#create a set of lattitude and longitude combinations
lats = np.random.uniform(-90.000, 90.000, 1500)
lngs = np.random.uniform(-180.000, 180.000, 1500)
lats_lngs = zip(lats, lngs)
lats_lngs
```

```python
#Add the latitudes and longitudes to a list
coordinates = list(lats_lngs)
coordinates
```

```python
#import citipy module which would help to find city based on lattitude and longitude
from citipy import citipy
```

```python
#iterate over the coordinates and find the nearest city using citipy
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1])
    print(city.city_name,",",city.country_code)
```

```python
#Create a unique list of cities
cities = []

#iterate over the coordinates and find nearest city and add it to the list if new
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    
    #find if the city name is not an existing one and add it to the list
    if city not in cities:
        cities.append(city)
    
#check the count of unique cities added
len(cities)
```

```python
import requests
from datetime import datetime
```

```python
requests.__version__
```

```python
from config import open_weather_api_key;
```

```python
#starting url for open weather api call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + open_weather_api_key
```

```python
#create empty city data list
city_data = []

#Print the beginning of the logging
print("Beginning of the Data Retrieval   ")
print("----------------------------------")

#create counters
record_counter = 1
set_counter = 1

#iterate through the cities
for i, city in enumerate(cities):
    
    #Group cities of 50 for logging purpose
    if (i % 50 == 0 and i >= 50):
        set_counter += 1
        record_counter = 1
    
    #Create endpoint url with each city
    city_url = url +"&q=" + city
    
    #log the city url, record counter and set counter
    print(f"Processing record {record_counter} of Set {set_counter} | {city}")
    #increment the record counter for this set
    record_counter += 1
    #make the api request to get weather data for each of the city
    try:
        #Parse the response and get json
        city_weather = requests.get(city_url).json()
        #Parse output and get the data needed
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_winds = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]

        #convert date to iso standard
        city_date = datetime.utcfromtimestamp(city_weather["dt"]).strftime('%Y-%m-%d %H:%M:%S')

        #app all the data to city list
        city_data.append({"City": city.title(),
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_winds,
                          "Country": city_country,
                          "Date": city_date})
    #Handle exception
    except:
        print(f"City not found. Skipping...")
        pass
#end of data loading
print("----------------------------------")
print("Data Retrieval is complete        ")
print("----------------------------------")
    
```

```python
#import linear regression from the SciPy stats module
from scipy.stats import linregress
```

```python
#function to create linear regression on the weather data and plot in the chart
#with linear regression line and equation 
def plot_linear_regression(x_values, y_values, title, y_label, text_coordinates):
    
    #calculate the regression for hemisphere data
    (slope, intercept, r_value, p_value, std_err) = linregress(x_values, y_values)
    
    print(r_value, p_value, std_err)
    #find regression line values to plot
    regress_values = x_values * slope + intercept;
    
    #create the linear regression formula
    line_eq = str.format(f"y = {str(round(slope, 2))}x + {str(round(intercept, 2))}")
    
    #create scatter plot with x and y values
    plt.scatter(x_values, y_values)
    
    #plot the linear regression line on the scatter chart
    plt.plot(x_values, regress_values, "r")
    
    #add annotation about the formula
    plt.annotate(line_eq, text_coordinates, fontsize=15, color="red")
    
    plt.xlabel('Latitude')

    plt.ylabel(y_label)
    
    plt.title(title)
    
    plt.show()
    
```

```python
#create northern hemisphere as a dataframe, if latitude is greater than 0 is northern hemisphere
northern_hemisphere_df = city_data_df[(city_data_df["Lat"] >=0)]
northern_hemisphere_df.head()
```

```python
#northern hemisphere get latitudes and max temperatures
northern_lats = northern_hemisphere_df["Lat"]
northern_max_temps = northern_hemisphere_df["Max Temp"]

#call the linear regression plotting function
plot_linear_regression(northern_lats, northern_max_temps, 
                       'Linear Regression on the Northern Hemisphere \n For Max Temperature',
                       'Max Temp', (0, 30))
```
