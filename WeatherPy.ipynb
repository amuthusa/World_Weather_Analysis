```python
#import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#create a set of lattitude and longitude combinations
lats = np.random.uniform(-90.000, 90.000, 1500)
lngs = np.random.uniform(-180.000, 180.000, 1500)
lats_lngs = zip(lats, lngs)
lats_lngs
```

```python
#Add the latitudes and longitudes to a list
coordinates = list(lats_lngs)
coordinates
```

```python
#import citipy module which would help to find city based on lattitude and longitude
from citipy import citipy
```

```python
#iterate over the coordinates and find the nearest city using citipy
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1])
    print(city.city_name,",",city.country_code)
```

```python
#Create a unique list of cities
cities = []

#iterate over the coordinates and find nearest city and add it to the list if new
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    
    #find if the city name is not an existing one and add it to the list
    if city not in cities:
        cities.append(city)
    
#check the count of unique cities added
len(cities)
```

```python
import requests
from datetime import datetime
```

```python
requests.__version__
```

```python
from config import open_weather_api_key;
```

```python
#starting url for open weather api call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + open_weather_api_key
```

```python
#create empty city data list
city_data = []

#Print the beginning of the logging
print("Beginning of the Data Retrieval   ")
print("----------------------------------")

#create counters
record_counter = 1
set_counter = 1

#iterate through the cities
for i, city in enumerate(cities):
    
    #Group cities of 50 for logging purpose
    if (i % 50 == 0 and i >= 50):
        set_counter += 1
        record_counter = 1
    
    #Create endpoint url with each city
    city_url = url +"&q=" + city
    
    #log the city url, record counter and set counter
    print(f"Processing record {record_counter} of Set {set_counter} | {city}")
    #increment the record counter for this set
    record_counter += 1
    #make the api request to get weather data for each of the city
    try:
        #Parse the response and get json
        city_weather = requests.get(city_url).json()
        #Parse output and get the data needed
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_winds = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]

        #convert date to iso standard
        city_date = datetime.utcfromtimestamp(city_weather["dt"]).strftime('%Y-%m-%d %H:%M:%S')

        #app all the data to city list
        city_data.append({"City": city.title(),
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_winds,
                          "Country": city_country,
                          "Date": city_date})
    #Handle exception
    except:
        print(f"City not found. Skipping...")
        pass
#end of data loading
print("----------------------------------")
print("Data Retrieval is complete        ")
print("----------------------------------")
    
```

```python
#create new dataframe from list of city data
city_data_df = pd.DataFrame(city_data)
city_data_df.head(10)
```

```python
#create new Column Order
new_column_order = ["City", "Country", "Date", "Lat", "Lng", "Max Temp", "Humidity", "Cloudiness", "Wind Speed"]
city_data_df = city_data_df[new_column_order]
```

```python
#create an outputfile and save the data
output_data_file = "weather_data/cities.csv"
#export weather data to csv file
city_data_df.to_csv(output_data_file, index_label="City_ID")
```

```python
#extract all the data needed for scatter plots
lats = city_data_df["Lat"]
max_temps = city_data_df["Max Temp"]
humidity = city_data_df["Humidity"]
cloudiness = city_data_df["Cloudiness"]
wind_speed = city_data_df["Wind Speed"]
```

```python
#import time module from python
import time

#Get today's date
today = time.strftime('%x')
today
```

```python
#Build the scatter plot between latitudes and max temperatures
plt.subplots(figsize=(12,8))
plt.scatter(lats, max_temps, marker="o",
           edgecolor="black", linewidths=1,
           alpha=0.8, label="Cities")
#set labels and title
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.title(f"City Latitude vs Max Temperature {today}")

#set grid
plt.grid(True)

#save figure
plt.savefig("weather_data/Fig1.png")

#show plot
plt.show()
```
